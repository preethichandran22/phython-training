```python
import openpyxl
wb=openpyxl.load_workbook('store.xlsx')
sheet=wb['Sheet1']
#cells=sheet.cell(2,3)
for row in range (2,sheet.max_row+1):
    qty=sheet.cell(row,7)
    unit_price=sheet.cell(row,8)
    value=qty.value * unit_price.value
    new_cell=sheet.cell(row,9)
    new_cell.value=value
    print(new_cell.value)
wb.save('store2.xlsx')
```
